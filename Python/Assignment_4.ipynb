```python
import requests,bs4,string
hostname='https://www.bizbaya.com'
path='/sme-india/rajasthan/jaipur'
fhand=requests.get(hostname+path)
```

```python
soup=bs4.BeautifulSoup(fhand.text,'html.parser')
ul_header=soup.find('div',{'class':'view-header'})
strong_tag=ul_header.find('strong')
tot_page=int(strong_tag.getText(separator=',').split(',')[2].split(':')[1].strip())
ul_tag=soup.find('table',{'class':'views-view-grid cols-1'})
links=[]
for a_tag in ul_tag.findAll('a'):
    links.append(a_tag.get('href'))
fhand.close()
# print(len(set(links)))
for page_no in range(1,tot_page):
    srclink=hostname+path+'?page='+str(page_no)
    fhand=requests.get(srclink)
    soup=bs4.BeautifulSoup(fhand.text,'html.parser')
    ul_tag=soup.find('table',{'class':'views-view-grid cols-1'})
    for a_tag in ul_tag.findAll('a'):
        links.append(a_tag.get('href'))
    fhand.close()
links=set(links)
# print(len(set(links)))
```

```python
import pandas as pd
dict_data={'SME Cpmpany Name':[], 'Director(s)':[], 'Mobile':[], 'Telephone':[], 'Fax':[], 'Email':[], 'Address':[], 'PIN Code':[], 'Products and Services':[]}
for link in links:
    fhand=requests.get(hostname+link)
    soup=bs4.BeautifulSoup(fhand.text,'html.parser')
    ul_tag=soup.find('div',{'class':'views-row views-row-1 views-row-odd views-row-first views-row-last'})
    if ul_tag is None:
        ul_tag=soup.find('div',{'class':'views-row views-row-1 views-row-odd views-row-first'})
        if ul_tag is None:
            print(hostname+link)
#     fhand.close()
#         print(len(ul_tag),hostname+link)
    temp_data={}        
    for p_tag in ul_tag.findAll('p'):
#         print(len(p_tag))
        key=p_tag.findChild().getText().strip().strip(':')
        value=p_tag.findChild().findNextSibling().getText().strip()
        value='na' if value is None else value
        if key in temp_data.keys():
            temp_data[key]=value
        else:
            temp_data[key]=value
    for key in dict_data.keys():
        if key in temp_data.keys():
            dict_data[key].append(temp_data[key])
        else:
            dict_data[key].append('na')
    fhand.close()
tempdf=pd.DataFrame(dict_data)                
tempdf.head()
```
