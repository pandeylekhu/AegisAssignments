```python
import nltk
from nltk.corpus import stopwords
from nltk.corpus import brown
```

```python
total_words=nltk.corpus.brown.words()
total_sentence=nltk.corpus.brown.sents()
print('total words in brown corpus: ', len(total_words))
print('total sentences in brown corpus: ', len(total_sentence))
```

```python
wordcount=len(total_words)
wordcount
```

```python
from string import punctuation
print (punctuation)
tokens=[w for w in total_words if w not in punctuation]
tokens[:20]
```

```python
bigrams=list(nltk.bigrams(tokens))
```

```python
bigrams
```

```python
unigram_freq=nltk.FreqDist(w.lower() for w in tokens)
unigram_freq
```

```python
bigram_freq=nltk.FreqDist(bigrams)
bigram_freq
```

```python
uni_prob={}
for i,j in unigram_freq.items():
#    print(i,j)
    unigram_probability=round((unigram_freq[j]/wordcount),3)
    uni_prob[i] = unigram_probability
```

```python
uni_prob
```

```python
bi_prob={}
for i,j in bigram_freq.items():
#    print(i,j)
    bigram_probability=round((bigram_freq[j]/wordcount),3)
    bi_prob[i] = bigram_probability
```

```python
bi_prob
```

```python
text='One day you will succeeed'
from nltk import word_tokenize
tokens = word_tokenize(text)

tokens
```

```python
sentence_probability=0
for w in tokens:
    if w in uni_prob.keys():
        sentence_probability+=uni_prob[w]
```
